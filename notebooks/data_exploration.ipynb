```python
# data_exploration.ipynb
import os
import matplotlib.pyplot as plt
from PIL import Image
from utils import list_images
```

```python
import os

PROJECT_ROOT = os.path.abspath("..")  # or the path to your project root
DATA_DIR = os.path.join(PROJECT_ROOT, "data", "processed", "images", "train")

print(DATA_DIR)
print(os.listdir(DATA_DIR))  # check if it works
```

```python
# Display class distribution
class_counts = {}
for cls in sorted(os.listdir(DATA_DIR)):
    cls_path = os.path.join(DATA_DIR, cls)
    if os.path.isdir(cls_path):
        class_counts[cls] = len(list_images(cls_path))

plt.bar(class_counts.keys(), class_counts.values())
plt.title("Number of images per class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
# Show sample images from each class
fig, axes = plt.subplots(2, 5, figsize=(15,6))
for i, cls in enumerate(sorted(class_counts.keys())):
    img_path = list_images(os.path.join(DATA_DIR, cls))[0]
    img = Image.open(img_path)
    ax = axes[i//5, i%5]
    ax.imshow(img, cmap='gray')
    ax.set_title(f"Class {cls}")
    ax.axis("off")
plt.show()
```
