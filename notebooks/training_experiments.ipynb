```python
# training_experiments.ipynb
import os
import matplotlib.pyplot as plt
import json
```

```python
# Plot training loss from checkpoints/log
CHECKPOINT_DIR = "models/checkpoints/"
log_file = os.path.join(CHECKPOINT_DIR, "train_log.json")  # optional if you saved logs
```

```python
# If you have stored train/val loss in a list
if os.path.exists(log_file):
    with open(log_file, "r") as f:
        logs = json.load(f)
    train_loss = logs["train_loss"]
    val_loss = logs["val_loss"]
else:
    # Example: dummy loss values for visualization
    train_loss = [0.1, 0.08, 0.06, 0.05]
    val_loss = [0.12, 0.09, 0.07, 0.06]
```

```python
plt.plot(train_loss, label="Train Loss")
plt.plot(val_loss, label="Validation Loss")
plt.title("Training Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
